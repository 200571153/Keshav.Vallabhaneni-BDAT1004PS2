```python
class Marsupial:
    def __init__(self):
        self.pouch = []

    def put_in_pouch(self, item):
        self.pouch.append(item)

    def pouch_contents(self):
        return self.pouch


class Kangaroo(Marsupial):
    def __init__(self, x, y):
        super().__init__()
        self.x = x
        self.y = y

    def jump(self, dx, dy):
        self.x += dx
        self.y += dy

    def __str__(self):
        return f'I am a Kangaroo located at coordinates ({self.x},{self.y})'
```

```python
# Test Marsupial
m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
print(m.pouch_contents())  
```

In this function, the base case for stopping the recursion is when x reaches 1. If x is even, the function is called with x/2, and if x is odd, the function is called with 3x + 1. The Collatz sequence continues until it reaches 1.

The code will print the Collatz sequence starting at the given positive integer x.

```python
def collatz(x):
    print(x)
    
    if x == 1:
        return
    elif x % 2 == 0:
        collatz(x // 2)
    else:
        collatz(3 * x + 1)
```

```python
collatz(1)
collatz(10)
```

Question 5

In this function, the recursion stops when n becomes less than or equal to 1. At each step of the recursion, the function divides n by 2 using integer division (//) and prints the remainder (the binary digit) using n % 2. The end='' argument in the print function is used to prevent newline characters after each digit so that they are printed together.

This code will correctly print the binary representation of the given non-negative integers.

```python
def binary(n):
    if n > 1:
        binary(n // 2)  # Recursively call binary with the integer division by 2
    print(n % 2, end='')
```

```python
binary(0)
print()
binary(1)
print()
binary(3) 
print()
binary(9)
```

This class defines a HeadingParser that uses the HTMLParser to parse the HTML content. It collects the heading data and stores it in a list with the heading level. The print_headings method is used to print the headings with appropriate indentation.


```python
from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.in_heading = False
        self.current_heading_level = 0
        self.headings = []

    def handle_starttag(self, tag, attrs):
        if tag.startswith("h") and tag[1:].isdigit():
            self.in_heading = True
            self.current_heading_level = int(tag[1])

    def handle_endtag(self, tag):
        if self.in_heading:
            self.in_heading = False

    def handle_data(self, data):
        if self.in_heading:
            self.headings.append((self.current_heading_level, data))

    def print_headings(self):
        for level, text in self.headings:
            indentation = ' ' * level
            print(indentation + text)
```

```python
infile = open('w3c.html', 'r', encoding='utf-8')
content = infile.read()
infile.close()
hp = HeadingParser()
hp.feed(content)
hp.print_headings()
```

In this code, we use the requests library to retrieve the web page's content, and BeautifulSoup to parse the HTML and extract links. The function recursively explores web pages up to the specified depth, and the indent variable is used to control the indentation level for each URL.


```python
import requests
from bs4 import BeautifulSoup

def webdir(url, depth, indent):
    if depth < 0:
        return

    # Create the indentation string
    indent_str = '  ' * indent

    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')

            # Print the current URL with appropriate indentation
            print(indent_str + url)

            if depth > 0:
                # Find and visit links on the current page
                links = soup.find_all('a')
                for link in links:
                    href = link.get('href')
                    if href.startswith('http'):
                        webdir(href, depth - 1, indent + 1)
    except requests.exceptions.RequestException:
        # Handle any request exceptions (e.g., invalid URLs)
        pass
```

```python
webdir('http://reed.cs.depaul.edu/lperkovic/csc242/test1.html', 2, 0)
```
